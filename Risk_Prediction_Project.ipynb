```python
import pandas as pd
url = "C:\\Users\\sanja\\Downloads\\Risk_Prediction\\train.csv"
data = pd.read_csv(url)
data.head(3)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score, roc_curve, roc_auc_score
```

```python
data1 = pd.get_dummies(data,drop_first = True)
data1.head(3)
x = data1.drop(["default payment next month"],axis = 1)
y = data1["default payment next month"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
model = DecisionTreeClassifier(random_state = 42,max_depth = 5)
model.fit(x_train,y_train)
```

```python
pred_y = model.predict(x_test)
```

```python
print("Acc using Decision :",accuracy_score(y_test,pred_y))
print("Recall using Decision :",recall_score(y_test,pred_y))
print("Prec using Decision :",precision_score(y_test,pred_y))
```

```python
#HYPER PARAMTER TUNING FOR DECISIONTREE CLASSIFIER
```

```python
param_grid = {"max_depth" : [None,3,5,10],"min_samples_split" : [2,5,10], "min_samples_leaf" : [1,2,3]}
g_s = GridSearchCV(estimator = model, param_grid = param_grid, scoring = "accuracy", cv = 5)
g_s.fit(x_train,y_train)
```

```python
g_s.best_params_
```

```python
best_model = g_s.best_estimator_
best_model
best_model.fit(x_train,y_train)
pred_y3 = best_model.predict(x_test)
```

```python
print("Acc using Decision after tuning :",accuracy_score(y_test,pred_y3))
print("Recall using Decision after tuning :",recall_score(y_test,pred_y3))
print("Prec using Decision after tuning :",precision_score(y_test,pred_y3))
```

```python
#RANDOM FOREST CLASSIFIER
```

```python
from sklearn.ensemble import RandomForestClassifier
x1 = data1.drop(["default payment next month"],axis = 1)
y1 = data1["default payment next month"]
x1_train,x1_test,y1_train,y1_test = train_test_split(x1,y1,test_size = 0.2,random_state = 42)
model2 = RandomForestClassifier(n_estimators = 50,random_state = 42)
model2.fit(x1_train,y1_train)
```

```python
pred1_y = model2.predict(x1_test)
```

```python
print("Acc using RandomForest :",accuracy_score(y1_test,pred1_y))
print("Recall using RandomForest :",recall_score(y1_test,pred1_y))
print("Pre using RandomForest :",precision_score(y1_test,pred1_y))
```

```python
#HYPERPARAMETER TUNING FOR RANDOMFORESTCLASSIFIER
```

```python
param_grid1 = {"n_estimators" :[50,60,70],"max_depth" : [None,10,20,30,40],"min_samples_split" : [2,5,10],"min_samples_leaf" : [1,5,10],"max_features" : ['auto', 'sqrt', 'log2']}
g_s_r = GridSearchCV(estimator = model2,param_grid = param_grid1,scoring = "accuracy",cv = 5)
g_s_r.fit(x1_train,y1_train) 
```

```python
g_s_r.best_params_
```

```python
best_model1  = g_s_r.best_estimator_
best_model1
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
x2 = data1.drop(["default payment next month"],axis = 1)
y2 = data1["default payment next month"]
x2_train,x2_test,y2_train,y2_test = train_test_split(x2,y2,test_size = 0.2,random_state = 42)
model3 = AdaBoostClassifier(n_estimators = 50,learning_rate = 1)
model3.fit(x2_train,y2_train)
```

```python
pred2_y = model3.predict(x2_test)
```

```python
print("Acc using AdaBoost :",accuracy_score(y2_test,pred2_y))
print("Recall using AdaBoost :",recall_score(y2_test,pred2_y))
print("Pre using AdaBoost :",precision_score(y2_test,pred2_y))
```

```python
#HYPERPARAMETER TUNINIG FOR ADABOOST CLASSIFIER
```

```python
param_grid2 = { "n_estimators" : [50,100,200],"learning_rate" : [0.01,0.1,1]}
g_s_a = GridSearchCV(estimator = model3,param_grid = param_grid2,scoring = "accuracy",cv = 5)
g_s_a.fit(x2_train,y2_train)
```

```python
g_s_a.best_params_
```

```python
best_est = g_s_a.best_estimator_
```

```python
pred2_y1 = best_est.predict(x2_test)
```

```python
print("Acc using AdaBoost after tuning :",accuracy_score(y2_test,pred2_y1))
print("Recall using AdaBoost after tuning:",recall_score(y2_test,pred2_y1))
print("Pre using AdaBoost after tuning:",precision_score(y2_test,pred2_y1))
```
